```python
#import important libraries

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

import pandas as pd
```

```python
#generate two simple texts as examples

text_1 = "data science is one of the most important courses in computer science"
text_2 = "this is one of the best data science courses"
text_3 = "the data scientists perform data analysis"
```

```python
#combine both texts to make a corpus

data_exp1 = [text_1, text_2, text_3]
data_exp1
```

```python
#create a reference object to the CountVectorizer constructor
#use the reference object to generate BoW matrix using the fit_transform() function

count_vectorizer = CountVectorizer()
c_vector_matrix_exp1 = count_vectorizer.fit_transform(data_exp1)
c_vector_matrix_exp1
```

```python
#use get_feature_names_out() to extract vocabulary

c_tokens_exp1 = count_vectorizer.get_feature_names_out()
c_tokens_exp1
```

```python
#convert BoW matrix to an array format

c_vector_matrix_exp1.toarray()
```

```python
#convert BoW matrix to a Pandas dataframe

df_c_exp1 = pd.DataFrame(data = c_vector_matrix_exp1.toarray(), columns = c_tokens_exp1)
df_c_exp1
```

```python
tf = df_c_exp1.div(df_c_exp1.sum(axis=1), axis=0)
tf
```

```python
# Calculate IDF

N = df_c_exp1.shape[0]
df = (df_c_exp1 > 0).sum(axis=0)
# idf formula
idf = np.log10(N / df) + 1
idf_tbl = pd.DataFrame(list(idf.items()), columns=['token', 'idf'])
idf_tbl
```

```python
#create a reference object to the TfidfVectorizer constructor
#use the reference object to generate tf.idf matrix using the fit_transform() function

tfidf_vect_exp1 = TfidfVectorizer()
t_tfidf_matrix_exp1 = tfidf_vect_exp1.fit_transform(data_exp1)
t_tfidf_matrix_exp1
```

```python
#use get_feature_names_out() to extract vocabulary

t_tokens_exp1 = tfidf_vect_exp1.get_feature_names_out()
t_tokens_exp1
```

```python
# Extracting IDF (Inverse Document Frequency) from the TfidfVectorizer matrix
idf_values_exp1 = tfidf_vect_exp1.idf_

# Creating a DataFrame for IDF values
df_idf_exp1 = pd.DataFrame(idf_values_exp1, index=t_tokens_exp1, columns=["IDF"])
df_idf_exp1
```

```python
#import scityblock (manhattan distance) from scipy

from scipy.spatial.distance import cityblock
```

```python
# #calculate scityblock (manhattan distance) between two document vectors (text1, text3)

distance_1_2 = cityblock(df_t_exp1.iloc[0], df_t_exp1.iloc[1])
distance_2_3 = cityblock(df_t_exp1.iloc[0], df_t_exp1.iloc[2])
distance_1_3 = cityblock(df_t_exp1.iloc[1], df_t_exp1.iloc[2])

print("Manhattan Distance between sentence 1 and sentence 2:", distance_1_2)
print("Manhattan Distance between sentence 2 and sentence 3:", distance_2_3)
print("Manhattan Distance between sentence 1 and sentence 3:", distance_1_3)
```

```python
#calculate percentage scityblock (manhattan distance)

print("Manhattan Distance (percentage) between sentence 1 and sentence 2:", (1/distance_1_2)*100,"%")
print("Manhattan Distance (percentage) between sentence 2 and sentence 3:", (1/distance_2_3)*100,"%")
print("Manhattan Distance (percentage) between sentence 1 and sentence 3:", (1/distance_1_3)*100,"%")
```
